```python
import pandas as pd

data = pd.read_csv('emails.csv')

print(data.head())
```

```python
import nltk
import pandas as pd
from nltk.util import ngrams
from nltk.probability import FreqDist
```

```python
nltk.download('punkt')
```

```python
def tokenize_text(text):
    tokens = nltk.word_tokenize(text)
    return tokens
```

```python
def extract_ngrams(tokens, n):
    n_grams = list(ngrams(tokens, n))
    return n_grams
```

```python
def calculate_ngram_frequency(n_grams):
    fdist = FreqDist(n_grams)
    return fdist
```

```python
ngram_data = []
for email_text in data['text']:
    tokens = tokenize_text(email_text)
    bigrams = extract_ngrams(tokens, 2)
    trigrams = extract_ngrams(tokens, 3)

    bigram_freq = calculate_ngram_frequency(bigrams)
    trigram_freq = calculate_ngram_frequency(trigrams)

    ngram_data.append({'Bigrams': bigram_freq, 'Trigrams': trigram_freq})
```

```python
print("Bigrams in the first email:")
print(ngram_data[0]['Bigrams'])

print("\nTrigrams in the first email:")
print(ngram_data[0]['Trigrams'])
```

```python
def calculate_ngram_distribution(tokens, n):
    ngrams_list = list(ngrams(tokens, n))
    ngram_freq_dist = FreqDist(ngrams_list)
    return ngram_freq_dist
```

```python
from nltk.util import ngrams
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
n_values = [4,3,2]
for n in n_values:
    ngram_freq_dist = calculate_ngram_distribution(tokens, n)
    print(f"Frequency distribution of {n}-grams:")
    print(ngram_freq_dist.most_common(10))
    ngram_freq_dist.plot(30, title=f'Frequency Distribution of {n}-grams')
    plt.show()
```

```python
from nltk.probability import ConditionalFreqDist
def calculate_conditional_frequencies(tokens, n):
    ngrams_list = list(ngrams(tokens, n))
    cfd = ConditionalFreqDist(ngrams_list)
    return cfd
```

```python
n_values = [2, 3]
for n in n_values:
    cfd = calculate_conditional_frequencies(tokens, n)

    print(f"Conditional frequencies of {n}-grams:")
    for condition in cfd.conditions():
        print(f"Condition: {condition}")
        cfd[condition].tabulate(10)
        print()
```

```python
n = 2
target_word = "your"
cfd = calculate_conditional_frequencies(tokens, n)
```

```python
if n == 1:
    if target_word in cfd:
        following_words = cfd[target_word]
        if following_words:
            print(f"Words commonly following '{target_word}':")
            following_words.tabulate(10)
        else:
            print(f"No words follow '{target_word}' in the dataset.")
    else:
        print(f"'{target_word}' not found in the dataset.")
else:
    if target_word in cfd:
        following_bigrams = cfd[target_word]
        if following_bigrams:
            print(f"Bigrams commonly following '{target_word}':")
            following_bigrams.tabulate(10)
        else:
            print(f"No bigrams follow '{target_word}' in the dataset.")
    else:
        print(f"'{target_word}' not found in the dataset.")
```

```python
def estimate_bigram_probability(bigram):
    condition, word = bigram[0], bigram[1]
    if condition in cfd:
        bigram_count = cfd[condition][word]
        condition_count = sum(cfd[condition].values())
        if condition_count > 0:
            probability = bigram_count / condition_count
            return probability
    return 0
target_bigram = ("Subject", ":")
probability = estimate_bigram_probability(target_bigram)
print(f"Probability of '{target_bigram[0]} {target_bigram[1]}': {probability:.4f}")
```
